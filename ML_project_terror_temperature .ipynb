```python
def create_dataset_month(terrorism, temcity, nday, temp, monterror):
    
    
    empty = []
    
    for w in range(1,nday,1):
        
        monDterror = monterror[monterror["iday"]==w]
        
        cities = {(i) for i in monDterror['city']}
        
        
        temp01 = temp.append(temcity.iloc[[w]])
      
        trans_tem01 = temp01.T
    
        tem_numpy01 = np.array(trans_tem01)
    
    
        list_numpy = [i for i in tem_numpy01]
    
        
        list1 = []
        list2 = []
    
        
        for i in list_numpy:
            if i[0] in cities:
                li = float(i[1:])
                
                list1.append(i[0])
                
                list2.append(li)
    
        dic = {"city":list1,
               "temperature":list2}
        
        df =pd.DataFrame(dic)
    
        new = pd.merge(monDterror,df)
        
        empty.append(new)

    terrorN = pd.concat([i for i in empty], sort = False)
    return terrorN
```

```python
def create_dataset_year(year):
    
    template1 = "C:/Users/Coral/Desktop/data/terrorism{Y}.CSV".format(Y = year)

    terrorism_ = load_data(template1)
    
    terrorism = terrorism_.fillna(0)
    
    template2 = "C:/Users/Coral/Desktop/data/temperature{Y}city.CSV".format(Y = year)

    temcity = load_data(template2)
    
    
    temp = temcity.iloc[[0]]
    
    if year%4:
        ndays = [32, 30, 32, 31, 32, 31, 32, 32, 31, 32, 31, 32]
    else:
        ndays = [32, 29, 32, 31, 32, 31, 32, 32, 31, 32, 31, 32]
        
        
    
    newEm = []
    
    for month in range(1,13,1):
        
        nday = ndays[month-1]
        
        monterror = terrorism[terrorism["imonth"]==month]
        
        f = create_dataset_month(terrorism, temcity, nday, temp, monterror)
        
        newEm.append(f)

    terrorismN = pd.concat([i for i in newEm], sort = False) 

    
    return terrorismN
    
    
    
```

```python
def save_dataset(year):
    data = create_dataset_year(year)
    
    return data.to_csv("C:/Users/Coral/Desktop/data/Base_data2016ML.CSV")
    
    
```

```python
from sklearn.ensemble import RandomForestRegressor

clf = RandomForestRegressor(max_features = 5, random_state = 0, n_estimators = 100)

```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(target_test, pred)

random_forest_rmse = np.sqrt(mse)

random_forest_rmse
```
